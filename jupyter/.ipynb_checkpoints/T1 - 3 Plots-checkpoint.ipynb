# Plot y visualizacion de los datos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import os
mainPath = "/home/user/projectAnacoda/python-course-ml/python-ml-course/datasets/"
filename = "customer-churn-model/Customer Churn Model.txt"
file = os.path.join(mainPath, filename)
```

```python
data = pd.read_csv(file)
```

```python
len(data)
```

```python
%matplotlib inline
#savefig(*.jpeg)
```

# Scatter Plot

```python
data.plot(kind="scatter", x="Day Mins", y="Day Charge")
```

```python
data.plot(kind="scatter", x="Night Mins", y="Night Charge")
```

```python
data.plot(kind="scatter", x="Day Mins", y="Night Charge")
```

```python
figure, axs = plt.subplots(2,2, sharey=True, sharex=True)
data.plot(kind="scatter", x="Day Mins", y="Day Charge", ax=axs[0][0])
data.plot(kind="scatter", x="Night Mins", y="Night Charge",ax=axs[0][1])
data.plot(kind="scatter", x="Day Calls", y="Day Charge", ax=axs[1][0])
data.plot(kind="scatter", x="Night Calls", y="Night Charge",ax=axs[1][1])
plt.savefig("/tmp/figure.png")
```

# Histograma

```python
plt.hist(data["Day Calls"], bins = 1 + int (np.ceil(np.log2(len(data)))))
plt.xlabel("Numero llama al dia")
plt.ylabel("Frecuencia")
plt.title("Frecuencia")
```
